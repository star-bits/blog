```python
# state-based vs. object-oriented

x = [1, 2, 3, 4, 5, 6, 7]

y1 = [0.5, 2, 4.5, 8, 12.5, 18, 24.5]
y2 = [4, 6, 7, 7, 6, 4, 1]
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(121)
plt.plot(x, y1, marker='o', ms=10, label="power")
plt.xlabel("time")
plt.ylabel("output")
plt.legend()

plt.subplot(122)
plt.plot(x, y2, ls=":", ms=10, label="torque")
plt.xlabel("time")
plt.legend()

plt.suptitle("performance")
plt.subplots_adjust(wspace=0.1)
plt.tight_layout()
```

```python
fig, axs = plt.subplots(ncols=2,
                        figsize=(8, 4),
                        gridspec_kw={"wspace":0.1},
                        constrained_layout=True)

axs[0].plot(x, y1, marker="o", ms=10, label="power")
axs[1].plot(x, y2, ls=":", ms=10, label="torque")

for ax in axs:
    ax.set_xlabel("time")
    ax.legend()
axs[0].set_ylabel("output")

fig.suptitle("performance")
```

- Figure: fig
    - Axes: axs[0], axs[1]
        - Artist: plot, labels, legend, suptitle
            - line: color, linestyle, linewidth, ...
            - patch: edgecolor, facecolor, linestyle, linewidth, ...
            - text: color, fontfamily, fontsize, fontstyle, fontweight, ...
